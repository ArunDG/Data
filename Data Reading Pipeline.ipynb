```python
import os
import pandas as pd

# Getting the current work directory (cwd)
thisdir = 'D:\\book_data_for_tableau_prep'

filelist=[]
def getfile(path="",typ=""):
    
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file.endswith(typ):
                file=os.path.join(r, file)
                filelist.append(file)
    DF=pd.DataFrame(filelist,columns=['filelist'])
    return DF

#Call the function to get list of files

getfile("D:\\NCS Projects\CANVAS\Development\Source to Staging\SRC\BBO",".csv")     
```
