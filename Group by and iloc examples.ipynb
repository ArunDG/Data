```python
# Group by Example

# Group data based on the single column
grouped=titanic.groupby('Survived')
type(grouped)

#Access values of the each  group
grouped.groups

#Select a specific group 
grouped.get_group(1)

#group data based on the multiple column 
titanic.groupby(['Survived','Sex']).size()

#Apply aggeregate function with group by
group=titanic.groupby(['Survived','Sex'])
group.agg({'SibSp':'size','Age':'mean','Fare':['sum','mean']})

# Apply aggeregate function with group by and named agg

Aggregation={
    'Cnt_SibSp':pd.NamedAgg(column='SibSp',aggfunc='size'),
    'Avg_Age':pd.NamedAgg(column='Age',aggfunc='mean'),
    'Tot_fare':pd.NamedAgg(column='Fare',aggfunc='sum'),
    'Avg_fare':pd.NamedAgg(column='Fare',aggfunc='mean')
}

# We can also apply shorter form of above aggrgation instead of specifying full expression
Aggregatio={
    'Cnt_SibSp':('SibSp','size'),
    'Avg_Age':('Age','mean'),
    'Tot_fare':('Fare','sum'),
    'Avg_fare':('Fare','mean')
}


df=group.agg(**Aggregatio).reset_index()
```

```python
#create some user id and random dates

random.seed(18)

random.seed(18)
signup_datetime = datetime.datetime.today()
signup_datetimes = []
total_signups = 60

for signup in range(total_signups):
    random_days = random.randint(1, 13)
    random_minutes = random.randint(1, 50)
    signup_datetime = signup_datetime + datetime.timedelta(days=random_days, minutes=random_minutes)
    signup_datetimes.append(signup_datetime)
```

```python
df1 = pd.DataFrame({'user_created_at': signup_datetimes, 'user_id': range(0, total_signups)})
df1.head()
```

```python
def calc_dt_field(df,col,new_col):
    df["year"+new_col] = pd.to_datetime(df[col]).dt.year
    return df
```

```python
df1=calc_dt_field(df1,"user_created_at","_user_created")
```
