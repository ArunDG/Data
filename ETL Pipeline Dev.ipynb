```python
# Import all control file and load into database
import pandas as pd
import os 
def load_cf(cf_loc,cf_name):
    cf_name=pd.read_csv(cf_loc)
    return cf_name
FileList=load_cf('D:\ETL Pipeline\Source Files\Control Files\FileList.csv','FileList')
AttributeFile=load_cf('D:\ETL Pipeline\Source Files\Control Files\AttributeFile.csv','AttributeFile')
AttributeFile.head(5)
```

```python
# combine atribute file and file list and create single attribute file and load into database

CTL_SRCFILE_ATTRIB=pd.merge(FileList,AttributeFile,how='inner',left_on='FileID',right_on='FileID')
```

```python
# extract filename and location of a source folder and loaded into DF

filelist=[]
def list_file(directory,extention):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(extention):
                filelist.append(directory + "\\" + file)
                #return filelist
            else:
                print ("There is no .txt file in the folder")
#create a DF from list of files
    if filelist:
        allfiles=pd.DataFrame(filelist,columns=['file_name'])
        return allfiles
    #else:
        #print("There is no files to load in the folder")
files=list_file('D:\ETL Pipeline\Source Files\Interface Files\Incoming\BBO','.txt')
```

```python
filelist
```

```python
#extract name key from file

files['fname']=pd.DataFrame([i for i in files['file_name'].str.split("\\",n=-1).str[-1:]])
files["Name_key"]= files["fname"].str[0:-19]
```

```python
#Combine attribute and file and  file list from the location to extract only relevent files
CTL_SRCFILE_ATTRIB.rename(columns={'Filename ':'Filename'},inplace=True)
full_list=pd.merge(CTL_SRCFILE_ATTRIB,files,how='inner',left_on='Filename',right_on='Name_key')
```
