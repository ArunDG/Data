```python
#convert char to numeric
ctn = pd.DataFrame({ 'product': ['A','B','C','D','E','F'], 
                   'price': ['10','20','30','40','80','120'],
                   'sales': ['20','-','60','-','80','80']
                  })
```

```python
# Import the sqlite3 module

import sqlite3

#create database connection in-memory database

connectionobject  =sqlite3.connect(":memory:")


# Obtain a cursor object

cusrsorobject    =connectionobject.cursor()


# Create a table in the in-memory database

createtable     ="CREATE TABLE EMP(ID INT,FIRSTNAME VARCHAR(32),LASTNAME VARCHAR(32))"

cusrsorobject.execute(createtable)

#print the tables 

cusrsorobject.execute("select * from SQLite_master where type=\"table\"")

print("Tables available in the in-memory database(main):")

tables =cusrsorobject.fetchall()

print("Listing tables from SQLite_master:")

for table in tables:

    print("------------------------------------------------------")

    print("DB Object Name: %s"%(table[0]))

    print("Name of the database object: %s"%(table[1]))

    print("Table Name: %s"%(table[2]))

    print("Root page: %s"%(table[3]))

    print("SQL statement: %s"%(table[4]))

    print("------------------------------------------------------")

connectionobject.close()
```
